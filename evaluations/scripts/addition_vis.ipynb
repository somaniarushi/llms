```python
import pandas as pd
```

```python
df = pd.read_csv(CSV)
df
```

```python
# For each length1 / length2, calculate average score and display as heatmap
# Group by length1, length2
grouped = df.groupby(["length1", "length2"])

# Calculate average score
average = grouped["score"].mean().unstack()

# Display as heatmap
import seaborn as sns

sns.heatmap(average, annot=True, fmt=".2f")
```
